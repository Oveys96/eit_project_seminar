```python
import matplotlib as mpl

# Set default plotting settings
mpl.rcParams.update(mpl.rcParamsDefault)

%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
# first example

sr = 44_100
t = np.linspace(0, 1, sr)
f = 4

# Set always a constant figure size
# Probably sometimes use subfigures:
# https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subfigures.html

plt.figure(figsize=(6, 3))

# Simple plot
plt.plot(t, np.cos(2 * np.pi * f * t), lw=3, label="φ=0")

# scatter plot
plt.scatter(
    t[::1000],  # x-axis
    np.cos(2 * np.pi * f * t[::1000] + np.pi / 2),  # y-axis
    lw=1,  # linewidth
    c="C1",  # Colorcode
    label="φ=π/2",  # label
)
# plt.stem()
# ...

plt.grid()  # plot grid()
plt.xlabel(r"$t$ in $s$")  # x label, use "$...$" for LaTeX equation style
plt.ylabel(r"$y(t)=A\cdot$cos$(2 π f t + φ)$")
plt.legend()  # plot legend
plt.tight_layout()  # tighthen the layout -> recommended

# save figures as vector graphic (.pdf, .svg)
# plt.savefig("images/two_cos.pdf")

# You need the plt.show() to visualise the plot in the .ipynb notebook
plt.show()
```

```python
# second example

x = np.linspace(1, 10000, 1000)

plt.figure(figsize=(6, 3))
plt.plot(x, x**2, lw=3, label="$\mathcal{O}(n^2)$")
plt.plot(x, x * np.log(x), lw=3, label="$\mathcal{O}(n\, \log{}\, n)$")

plt.ylabel("$\mathcal{O}$")
plt.xlabel("$n$")

plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('images/big_o_fft.pdf')
plt.show()
```
